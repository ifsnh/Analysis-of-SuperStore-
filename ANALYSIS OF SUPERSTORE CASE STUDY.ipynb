### PROFIT & LOSS DATA SET'S of Region's & Discount's

```python
loss = d[d['Profit']<0]
profit = d[d['Profit']>0]
```

```python
profit
```

```python
a = profit['Profit'].mean()
b = profit['Profit'].sum()
c = profit['Discount'].mean()
print("The average profit in profit dataset is {} and total profit is {}".format(a,b))
print("The average discount is {}".format(c*100))
```

```python
loss
```

```python
a = loss['Profit'].mean()
b = loss['Profit'].sum()
c = loss['Discount'].mean()
print("The average profit in loss dataset is {} and total profit is {}".format(a,b))
print("The average discount is {}".format(c*100))
```

```python
x = sns.countplot(data = loss, x = 'Region')
for i in x.containers:
    x.bar_label(i,)
```

```python
plt.figure(figsize=(15,4))
plt.subplots_adjust(wspace=0.33)
plt.subplot(1,3,1)
plt.title('Total loss by region')
x = sns.barplot(data = loss, x= 'Region',y='Profit',estimator=np.sum,palette = "light:#5A9")
for i in x.containers:
    x.bar_label(i,)
plt.subplot(1,3,2)
plt.title('Average loss by region')
y = sns.barplot(data = loss, x= 'Region',y='Profit', palette = "flare")
for i in y.containers:
    y.bar_label(i,)
```

```python
gdsloss = loss.groupby(["Region","State"]).agg({'Sales': 'sum', 'Profit': 'sum','Discount': 'mean'})
print(gdsloss)
```

```python
sns.lineplot(data = gdsloss, x = "Discount", y = 'Profit', color= "b",marker = "p")
```

```python
plt.figure(figsize=(30,10))
sns.lineplot(data = gdsloss, x = "State", y = "Profit", marker= ".", dashes=True,color="r",ms =20)
```

```python
fig, ax = plt.subplots()
ax= sns.lineplot(x= 'Discount', y='Profit', data=loss, color="r", legend = True)
ax= sns.lineplot(x= 'Discount', y='Profit', data=profit, color="g", legend = True)
```

```python
fig, axes = plt.subplots(2,3,figsize=(25,20))
axes[0,0].set_title("Total Data on Profit")
x = sns.barplot(data=d, x='Ship Mode', y = 'Profit',hue = "Region", estimator = np.sum, ax=axes[0,0])
for i in x.containers:
    x.bar_label(i,)
axes[0,1].set_title("Loss Data on Profit")
y = sns.barplot(data=d, x='Ship Mode', y = 'Profit',hue = "Region", estimator = np.sum, ax=axes[0,1])
for i in y.containers:
    y.bar_label(i,)
axes[0,2].set_title("Profit Data on Profit")
z = sns.barplot(data=d, x='Ship Mode', y = 'Profit',hue = "Region", estimator = np.sum, ax=axes[0,2])
for i in z.containers:
    z.bar_label(i,)
axes[1,0].set_title("Total Data on Discount")
x = sns.barplot(data=d, x='Ship Mode', y = 'Discount',hue = "Region", estimator = np.mean, ax=axes[1,0])
for i in x.containers:
    x.bar_label(i,)
axes[1,1].set_title("Loss Data on Discount")
y = sns.barplot(data=d, x='Ship Mode', y = 'Discount',hue = "Region", estimator = np.mean, ax=axes[1,1])
for i in y.containers:
    y.bar_label(i,)
axes[1,2].set_title("Profit Data on Discount")
z = sns.barplot(data=d, x='Ship Mode', y = 'Discount',hue = "Region", estimator = np.mean, ax=axes[1,2])
for i in z.containers:
    z.bar_label(i,)
```
